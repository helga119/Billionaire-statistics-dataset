```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('world_population.csv')
data.head()
```

```python
columns_data_types = data.dtypes


print(columns_data_types)
```

```python
## Forward fill missing values
data.fillna(method='ffill', inplace=True) 
# drop rows with any remaining missing values
data.dropna(inplace=True) 
data.drop_duplicates(inplace=True)
```

```python
# selects the numeric columns
numeric_data = data.select_dtypes(include=['float64', 'int64'])


print("Summary Statistics (for Numeric Columns):")
print(numeric_data.describe())

```

```python
print("Correlation Matrix (for Numeric Columns):")
print(numeric_data.corr())



```

```python
def plot_population_growth(country, data):
    
    """
    Plots the population growth of a specified country over time using a line plot.
    
    Parameters:
    - country: String, The name of the country to plot.
    - data: DataFrame, A pandas DataFrame containing the population data.
    
    """
    country_data = data[data['Country/Territory'] == country]
    
    populations = [
        country_data['1970 Population'].values[0], 
        country_data['1980 Population'].values[0],
        country_data['1990 Population'].values[0],
        country_data['2000 Population'].values[0],
        country_data['2010 Population'].values[0],
        country_data['2015 Population'].values[0],
        country_data['2020 Population'].values[0],
        country_data['2022 Population'].values[0]
    ]
    
    years = [1970, 1980, 1990, 2000, 2010, 2015, 2020, 2022]
    
    plt.figure(figsize=(10, 5))
    plt.plot(years, populations, marker='o', linestyle='-', color='blue')
    plt.title(f'Population Growth Over Time for {country}')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.grid(True)
    plt.show()



```

```python
plot_population_growth('Nigeria', data)
```

```python
plot_population_growth('Niger', data)
```

```python
def plot_population_by_continent(data):
    """
    Creates a bar chart comparing the population distribution across continents.
    
    Parameters:
    - data: DataFrame, a pandas DataFrame containing the population data and continent information.
    """

    population_by_continent = data.groupby('Continent')['2022 Population'].sum()
    
    fig, ax = plt.subplots(figsize=(10, 5), dpi=144)
    ax.bar(population_by_continent.index, population_by_continent.values, color='skyblue')
    ax.set_title('Population Distribution by Continent in 2022')
    ax.set_xlabel('Continent')
    ax.set_ylabel('Total Population')
    ax.grid(True)
    plt.show()

```

```python
plot_population_by_continent(data)
```

```python
def plot_correlation_heatmap(data):
    """
    Plots a heatmap to illustrate the correlation between population size, area, and density.
    
    Parameters:
    - data: DataFrame, a pandas DataFrame containing the population, area, and density data.
    
    """
    
    #selected columns for correlation
    columns = ['2022 Population', 'Area (km²)', 'Density (per km²)']
    correlation_data = data[columns].corr()
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_data, annot=True, fmt=".2f", cmap='RdBu')
    plt.title('Correlation Heatmap')
    plt.show()

```

```python
plot_correlation_heatmap(data)
```
